```python
# Dependencies
from bs4 import BeautifulSoup
from splinter import Browser
import pymongo
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# URL of page to be scraped
url = 'https://www.surfline.com/surf-reports-forecasts-cams-map/@9.595333662946327,-84.298095703125,7z'

browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
#home = soup.select("div.sl-spot-list__spots")
locations = []
location = {}
results = soup.find_all('div', class_='sl-spot-list__ref')
for res in results:
    if len(res["class"]) != 1:
        continue
    try:
        # Identify and return title of town
        location["name"] = res.find('h3', 'sl-spot-details__name').text
        location["url"] = "https://www.surfline.com" + str(res.a["href"])
        location["surfheight"] = res.find('span', 'quiver-surf-height').text
        locations.append(location)
        location={}
        print(locations)
    except Exception as e:
        print(e)
```

```python
for location in locations:
    url = location["url"]
    browser.visit(url)
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    results = soup.find_all('div', class_='sl-wetsuit-recommender__weather')
    for res in results:
        try:
            location['waterTemp'] = res.find_next('div').text
            location['airTemp'] = res.find_next('div').find_next("div").text
        except Exception as e:
            print(e)
#print(locations)
```

```python
print(locations)
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
# Define database and collection
db = client.surfDB
collection = db.surf_summary
for location in locations:
    collection.insert_one(location)
```
